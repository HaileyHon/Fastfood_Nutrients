```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/haileyhon/Desktop/Fastfood.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df_rename = df.rename(
    columns = {
        'Resturant' : 'restaurants',
        'Item' : 'items',
        'Calories' : 'calories', 
        'Calories from\nFat' : 'calories_from_fat', 
        'Total Fat\n(g)' : 'total_fat(g)', 
        'Saturated Fat\n(g)' : 'saturated_fat(g)', 
        'Trans Fat\n(g)' : 'trans_fat(g)',
        'Cholesterol\n(mg)' : 'cholesterol(mg)',
        'Sodium \n(mg)' : 'sodium(mg)', 
        'Carbs\n(g)' : 'carbohydrate(g)',
        'Fiber\n(g)' : 'fiber(g)',
        'Sugars\n(g)': 'sugar(g)',
        'Protein\n(g)' : 'protein(g)'
        }
    )
df_rename.head()
```

```python
pd.set_option('display.max_rows', None)
```

```python
df_rename
```

```python
print(df_rename.isnull().mean())
```

```python
df_rename.dropna(subset = ['fiber(g)'], inplace = True)

df_rename.dropna(subset = ['protein(g)'], inplace = True)

print(df_rename.isnull().mean())
```

```python
df_rename.dtypes
```

```python
for col, dtype in df_rename.dtypes.items():
    if dtype == object:
        df_rename[col] = df_rename[col].str.strip()
```

```python
df_rename['restaurants'].unique()
```

```python
replace_names = {
    "Mcdonalds": "McDonald's",
    "McDonald’s": "McDonald's",
    "Wendy’s": "Wendy's",
}

df_rename["restaurants"] = df_rename["restaurants"].replace(replace_names)

# df_rename["restaurants"] = df_rename["restaurants"].replace(["Mcdonalds","McDonald’s"], "McDonald's")
# df_rename["restaurants"] = df_rename["restaurants"].replace("Wendy’s", "Wendy's")

df_rename['restaurants'].unique()
```

```python
duplicates = df_rename[df.duplicated()] 
print(duplicates)
```

```python
df_rename.drop_duplicates(keep=False, inplace=True)

duplicates = df_rename[df.duplicated()] 
print(duplicates)
```

```python
restaurants = df_rename["restaurants"].unique()
for restaurant in restaurants:
    df_restaurant = df_rename.loc[df_rename["restaurants"] == restaurant]
    df_restaurant_fiber = df_restaurant.sort_values("fiber(g)", ascending=False).head(5)
    df_restaurant_fiber.plot.bar(x="items", y="fiber(g)", title=f'{restaurant}', color="green")
```

```python
restaurants = df_rename["restaurants"].unique()
for restaurant in restaurants:
    df_restaurant = df_rename.loc[df_rename["restaurants"] == restaurant]
    df_restaurant_sodium = df_restaurant.sort_values("sodium(mg)", ascending=False).head(5)
    df_restaurant_sodium.plot.bar(x="items", y="sodium(mg)", title=f'{restaurant}', color="grey")
```

```python
df_fiber_max = df_rename.sort_values('fiber(g)', ascending=False).drop_duplicates('restaurants')[["restaurants", "items", "fiber(g)"]]
df_fiber_bar = df_fiber_max.plot.bar(x="restaurants", y="fiber(g)", color="green")

print(df_fiber_max)
```
